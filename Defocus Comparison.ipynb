```python
%matplotlib notebook
import numpy as np
import poppy
import matplotlib.pyplot as plt
import astropy.units as u
from llowfs import make_coronagraph
```

```python
oversample=4
fov = 0.5 #arcsec
npix_sensor = 128
pixelscale = fov/npix_sensor  #arcsec/pixel
wavelength = 0.632e-6*u.m
sensor_defocus = 0.5 #(times wavelength)
charge=2
```

```python
llowfs_05 = make_coronagraph([0,0,0,0],wavelength=wavelength,oversample=2,pixelscale=pixelscale,\
                                sensor_defocus=0.5,llowfs=True,mask_type='vortex',vortex_charge=charge)
llowfs_1 = make_coronagraph([0,0,0,0],wavelength=wavelength,oversample=2,pixelscale=pixelscale,\
                                sensor_defocus=1,llowfs=True,mask_type='vortex',vortex_charge=charge)
llowfs_2 = make_coronagraph([0,0,0,0],wavelength=wavelength,oversample=2,pixelscale=pixelscale,\
                                sensor_defocus=2,llowfs=True,mask_type='vortex',vortex_charge=charge)
```

```python
psf_05 = llowfs_05.calc_psf(wavelength=wavelength, display_intermediates=False)
psf_1 = llowfs_1.calc_psf(wavelength=wavelength, display_intermediates=False)
psf_2 = llowfs_2.calc_psf(wavelength=wavelength, display_intermediates=False)
```

```python
plt.figure()
poppy.display_psf_difference(psf_05,psf_2,title='LLOWFS Response (Flat WF)')
```

```python
plt.figure()
poppy.display_psf_difference(psf_1,psf_2,title='LLOWFS Response (Flat WF)')
```

```python
plt.figure()
poppy.display_psf_difference(psf_05,psf_1,title='LLOWFS Response (Flat WF)')
```
