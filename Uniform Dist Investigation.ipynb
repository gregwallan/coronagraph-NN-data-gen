# Results for -100 to 100 nm

```python
highest_coeff = 15
M = highest_coeff-1

lam = 632

'''
with h5py.File('uniform_recons_03302020.mat', 'r') as f:
#with h5py.File('recon_03242020.mat', 'r') as f:
    truth948nm = np.array(f["test_output"])*10
    print(truth948nm.shape)
    rec948nm = np.array(f['rec_test_128px'])*10
'''

#4/5/20
with h5py.File('reconstructions_04052020.mat', 'r') as f:
    truth948nm = np.array(f["test_output_dataset_126nm"])*10
    print(truth948nm.shape)
    rec948nm = np.array(f['rec_test_dataset_126nm'])*10


print(rec948nm.shape)
print(rec948nm[:10,0])
```

```python
colormap = plt.cm.nipy_spectral
colorcycle = [colormap(c) for c in np.linspace(0, 1,M)]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_prop_cycle(color=colorcycle)

for ii in range(M):
    ax.hist(truth948nm[:,ii],bins=50,histtype='step')
    
ax.set_xlabel('True Zernike Value (nm)')
ax.set_ylabel('Occurences')
```

```python
overall = np.sqrt(np.sum(np.power(truth948nm,2),axis=1))
plt.figure()
plt.hist(overall,bins=50)
plt.xlabel('Overall WFE Value (nm)')
plt.ylabel('Occurences')
```

```python
for ii in range(M):
    plt.figure()
    plt.plot([-948,948],[-948,948],"--",color='grey')
    plt.plot(truth948nm[:,ii],rec948nm[:,ii],'.')
    plt.title('Z'+str(ii+2))
    plt.xlabel('True Value (nm)')
    plt.ylabel('Measured Value (nm)')
```
