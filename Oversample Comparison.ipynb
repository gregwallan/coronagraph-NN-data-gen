```python
%matplotlib notebook
import numpy as np
import poppy
import matplotlib.pyplot as plt
import astropy.units as u
from llowfs import make_coronagraph
```

```python
oversample=4
fov = 0.5 #arcsec
npix_sensor = 128
pixelscale = fov/npix_sensor  #arcsec/pixel
wavelength = 0.632e-6*u.m
sensor_defocus = 0.5 #(times wavelength)
charge=2
```

```python
llowfs_2 = make_coronagraph([0,0,0,0],wavelength=wavelength,oversample=2,pixelscale=pixelscale,\
                                sensor_defocus=sensor_defocus,llowfs=True,mask_type='vortex',vortex_charge=charge)
llowfs_4 = make_coronagraph([0,0,0,0],wavelength=wavelength,oversample=4,pixelscale=pixelscale,\
                                sensor_defocus=sensor_defocus,llowfs=True,mask_type='vortex',vortex_charge=charge)
llowfs_8 = make_coronagraph([0,0,0,0],wavelength=wavelength,oversample=8,pixelscale=pixelscale,\
                                sensor_defocus=sensor_defocus,llowfs=True,mask_type='vortex',vortex_charge=charge)
```

```python
psf_2 = llowfs_2.calc_psf(wavelength=wavelength, display_intermediates=False)
```

```python
psf_4 = llowfs_4.calc_psf(wavelength=wavelength, display_intermediates=False)
psf_8 = llowfs_8.calc_psf(wavelength=wavelength, display_intermediates=False)
```

```python
plt.figure()
poppy.display_psf_difference(psf_2,psf_8,title='LLOWFS Response (Flat WF)')
```

```python
plt.figure()
poppy.display_psf_difference(psf_2,psf_4,title='LLOWFS Response (Flat WF)')
```

```python
plt.figure()
poppy.display_psf_difference(psf_4,psf_8,title='LLOWFS Response (Flat WF)')
```
